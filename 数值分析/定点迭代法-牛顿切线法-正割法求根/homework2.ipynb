```python
#用定点迭代法求x = sqrt(2*x(i))在1.2附近的根
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0, 5, 0.1)
y = x
y2 = (2*x)**0.5
plt.plot(x, y)
plt.plot(x, y2)
x1 = 0.2
def f(x):
    #X(i+1) = sqrt(2*X(i))
    return (2*x)**0.5
y1 = x1
y2 = f(x1)
#直到误差小于十的负三次方停止迭代
while abs(y2-y1)>10**(-3):
    x2 = y2
    y1 = x2
    y2 = f(x2)
    plt.plot([x2,y1],[x2,y2])
    plt.plot([x2,y2],[y2,y2])
    print(x2)
print(x2)
```

```python
#用牛顿切线法求x = 3*np.exp(x)-4*np.cos(x)在1附近的根
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0, 1, 0.1)
y = 3*np.exp(x)-4*np.cos(x)
plt.grid()
plt.plot(x, y)
x1 = 1
def f(x):
    return 3*np.exp(x)-4*np.cos(x)
def g(x):
    return 3*np.exp(x)+4*np.sin(x)
x2 = x1 - f(x1)/g(x1) #由直线的坐标和斜率可得该直线与x轴的交点
plt.plot([x1,x1],[0,f(x1)])
plt.plot([x1,x2],[f(x1),0])
print(x2)
while (x1-x2)>10**-3: #直到相邻两交点的值小于某一规定值
    x3 = x2 - f(x2)/g(x2) #求下一交点的值
    plt.plot([x2,x3],[f(x2),0])
    x1 = x2 #将原先的x2点作为为x1
    x2 = x3 #将下一交点的值x3作为x2
    plt.plot([x1,x1],[0,f(x1)])
    print(x2)
print(x2)
```

```python
#用正割法求x = 3*np.exp(x)-4*np.cos(x)在1附近的根
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0, 1, 0.1)
y = 3*np.exp(x)-4*np.cos(x)
plt.grid()
plt.plot(x, y)
x1 = 1 #初始两点选择1和0.8
x2 = 0.8
def f(x):
    return 3*np.exp(x)-4*np.cos(x)
x3 = x2-((f(x2)*(x1-x2))/(f(x1)-f(x2)))#由两点式求出直线与x轴交点
while (x1-x2)>10**(-3): #直到相邻两交点的值小于某一规定值
    x3 = x2-((f(x2)*(x1-x2))/(f(x1)-f(x2))) #递推下一交点
    plt.plot([x1,x1],[0,f(x1)])
    plt.plot([x2,x2],[0,f(x2)])
    plt.plot([x1,x3],[f(x1),0])
    x1 = x2 #将原先的x2点作为为x1
    x2 = x3 #将下一交点的值x3作为x2
    print(x2)
```
